```python
speech_folder = './speeches'
model_name = "124M"
directory = os.listdir(speech_folder)[1:]
```

```python
sess.close()
tf.reset_default_graph()
```

```python
for filename in directory:
    path = speech_folder + '/' + filename
    run_name = filename.replace('.txt', '')
    print('-----------------------------------------------------------')
    print(f'**training on {filename}**')
    sess = gpt2.start_tf_sess()
    print(f'**checkpoint saved as {run_name}**')
    gpt2.finetune(sess,
                  path,
                  print_every=100,
                  #restore_from='fresh',
                  batch_size = 1,
                  model_name=model_name,
                  run_name = run_name,
                  steps=300)
    print(f'**end of training on {filename}**')
    print('-----------------------------------------------------------')
    
    sess.close()
    tf.reset_default_graph()
```

```python
for run_name in os.listdir('./checkpoint'):
    sess = gpt2.start_tf_sess()
    gpt2.load_gpt2(sess, run_name = run_name)
    result = gpt2.generate(sess, run_name=run_name, prefix = 'My fellow Americans,', length = 150, return_as_list=True)
    speaker = run_name
    results[speaker] = result
    sess.close()
    tf.reset_default_graph()
```
